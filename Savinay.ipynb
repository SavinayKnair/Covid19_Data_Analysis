### task 3.2: find maxmimum infection rate for China

```python
corona_dataset_aggregated.loc['China'].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []

# Loop through each country and calculate the max infection rate
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())

# Create a new DataFrame with the 'max_infection_rate' column
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

### Task 3.4: create a new dataframe with only needed column 

```python

corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
```

```python
corona_data.head()
```

### Task 5.1: Plotting GDP vs maximum Infection rate

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming corona_data is already defined and contains the relevant columns
x = corona_data["GDP per capita"]
y = corona_data["max_infection_rate"]

# Create the scatter plot
sns.scatterplot(x=x, y=np.log(y))  # Pass x and y as keyword arguments
plt.title("Scatter Plot of GDP per Capita vs Log of Max Infection Rate")
plt.xlabel("GDP per Capita")
plt.ylabel("Log of Max Infection Rate")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming corona_data is already defined and contains the relevant columns
x = corona_data["GDP per capita"]
y = corona_data["max_infection_rate"]

# Create the regression plot
sns.regplot(x=x, y=np.log(y))  # Pass x and y as keyword arguments
plt.title("Regression Plot of GDP per Capita vs Log of Max Infection Rate")
plt.xlabel("GDP per Capita")
plt.ylabel("Log of Max Infection Rate")
plt.show()
```
